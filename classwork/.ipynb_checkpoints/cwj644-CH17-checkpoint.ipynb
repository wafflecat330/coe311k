```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the parameters
k = 1.0        # Constant in the equation (e.g., spring constant)
h = [.01, .02, .03, .05, .1, .2, .3, .5]       # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx
```

```python
# Euler's Method loop

plt.figure(figsize=(10, 5))
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.grid(True)

for j in range(len(h)):
    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h[j], h[j])
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0
    
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    
        # Update equations
        y_values[i] = y_n + h[j] * v_n
        v_values[i] = v_n + h[j] * (-k * y_n)
        
    # Plot the results
    plt.plot(x_values, y_values, label=f'y, h={h[j]}')
    plt.plot(x_values, v_values, label=f'v, h={h[j]}', linestyle='--')
    plt.legend()
```

Write-up: Increasing h causes the oscillation to slowly develop a negative damping effect. The higher that h is, the sooner and more extreme the deviation from the original amplitude is.
