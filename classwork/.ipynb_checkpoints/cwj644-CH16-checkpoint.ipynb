```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(h):
    x_start, x_end = 0, 10
    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1 # y(0) = 1

    # Function representing dy/dx = -2y
    def f(x, y):
        return -2 * y

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    
    
    return x_values, y_values
```

```python
def exact_solution(x):
    return np.exp(-2 * x)
```

```python
def error(h_values):
    
    for i in range(len(h_values)):
        x_values, y_approx = euler(h_values[i])
        y_exact = exact_solution(x_values)
        
        global_error = np.sum(abs(y_exact - y_approx))
        print(f"h = {h_values[i]:.1f} ... Global Error: {global_error}")
        
        local_error = abs(y_exact - y_approx)
        
        # Plot the error
        plt.plot(x_values, local_error, label=f"h={h_values[i]}")
        plt.xlabel('x')
        plt.ylabel('Error')
        plt.title("Error of euler's method")
        plt.legend()
        plt.show()
```

```python
h_values = np.array([0.1, 0.5, 1.0, 1.5])
global_error = np.zeros(4)

error(h_values)
```

As step size h increases, the accumulated (global) and error for each point (local) increases because derivative approximations become less accurate.
