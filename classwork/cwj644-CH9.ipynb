## Part 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Data points 

X = np.array([1, 2, 3, 4, 5])

Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.figure()
plt.scatter(X, Y)

x_lstsq = np.arange(1, 5.01, .01)
y_lstsq = x_lstsq * m + c
plt.plot(x_lstsq, y_lstsq, 'r-')
```

## Part 2

```python
# Data points 

rng = np.random.default_rng()

X2 = rng.random(5) * 5
X2 = np.sort(X2)

Y2 = rng.random(5) * 5

# Stack the design matrix with a column of ones (for the intercept)

A2 = np.vstack([X2, np.ones(len(X2))]).T

# Solve using least squares

m2, c2 = np.linalg.lstsq(A2, Y2, rcond=None)[0]

print(f"Slope: {m2}, Intercept: {c2}")
```

```python
plt.figure()
plt.scatter(X2, Y2)

x2_lstsq = np.linspace(X2.min(), X2.max()+0.01, 100)
y2_lstsq = x2_lstsq * m2 + c2
plt.plot(x2_lstsq, y2_lstsq, 'r-')
```
