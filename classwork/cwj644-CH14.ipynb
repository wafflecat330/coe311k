```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi  # Interval
N = 50  # Number of sample points

# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Plot the sampled points
plt.plot(x, y, 'o-', label='Sampled Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Uniform Sampling of f(x) = sin(x)')
plt.legend()
plt.show()
```

```python
# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx_c = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
df_dx_f = (f(x+h) - f(x)) / h

# Backward Difference Approximation
df_dx_b = (f(x) - (f(x-h))) / h

# Plot the approximation and the true derivative
plt.plot(x, np.cos(x), label='Exact Derivative')
plt.plot(x, df_dx_c, 'o--', label='Central Difference Approximation')
plt.plot(x, df_dx_f, 'r--', label='Forward Difference Approximation')
plt.plot(x, df_dx_b, 'g--', label='Backward Difference Approximation')
plt.plot()
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Approximating Derivative of f(x) = sin(x) using Central Difference")
plt.legend()
plt.show()
```

Looking at the graph, we can see that the forward difference approximation is slightly behind the derivative and backward difference is slightly ahead. Central difference seems to be the closest of the three approximations
