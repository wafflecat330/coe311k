```python
import numpy as np
import math
import random

def approximatePi(numSamples):
    counter = 0
    for s in range(numSamples):
        x = np.random.uniform(0, 2) - 1
        y = np.random.uniform(0, 2) - 1
        
        if (x**2+y**2 < 1):
            counter = counter + 1
    
    return 4.0 * counter / numSamples
```

```python
for i in range(0, 100):
    calculated_pi = approximatePi(1000)
    true_pi = np.pi
    true_error = abs(true_pi - calculated_pi)
    relative_error = abs(true_pi - calculated_pi)/true_pi
    
    print(calculated_pi, true_pi, relative_error)
```
