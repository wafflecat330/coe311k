```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lstsq(X, Y):
    n = len(X)
    sum_X = np.sum(X)
    sum_Y = np.sum(Y)
    sum_XY = np.sum(X*Y)
    sum_XX = np.sum(X**2)
    
    m = (n*sum_XY - sum_X*sum_Y) / (n*sum_XX - (sum_X)**2)
    c = (sum_Y - m*sum_X) / n
    
    return m, c
```

This algorithm takes the system of normal equations and rearranges to solve for m and c.

```python
# Data points 

rng = np.random.default_rng()

X = rng.random(5) * 5
X = np.sort(X)

Y = rng.random(5) * 5

# Solve using least squares

m, c = lstsq(X, Y)

print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.figure()
plt.scatter(X, Y)

X_lstsq = np.linspace(X.min(), X.max(), 100)
Y_lstsq = X_lstsq * m + c
plt.plot(X_lstsq, Y_lstsq, 'r-')
```

```python
X2 = np.arange(0, 1.01, .01)
Y2 = np.pi*X2 + 7 + rng.random(len(X2))*2 - 1

m2, c2 = lstsq(X2, Y2)

print(f'm: {m2}  c: {c2}')
```

```python
plt.figure()
plt.scatter(X2, Y2)

Y2_lstsq = X2 * m2 + c2
plt.plot(X2, Y2_lstsq, 'r-')
```

```python
residuals = abs(Y2_lstsq - Y2)
residual_avg = np.mean(residuals)
residual_min = np.min(residuals)
residual_max = np.max(residuals)
residual_squared_sum = np.sum(residuals**2)

print(f'Residual Average: {residual_avg}')
print(f'Residual Minimum: {residual_min}')
print(f'Residual Maximum: {residual_max}')
print(f'Sum of Squared Residuals: {residual_squared_sum}')
print(f'Residuals: {residuals}')
```
