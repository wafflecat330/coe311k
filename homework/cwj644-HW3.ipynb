```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def taylorSinApprox(x, numTerms):
    output = 0
    for n in range(numTerms):
        output = output + (((-1)**n)*(x**(2*n+1)))/(math.factorial(2*n+1))
    return output
```

```python
x_range = np.arange(0,10.1,0.1)

approx_one = taylorSinApprox(x_range, 1)
approx_two = taylorSinApprox(x_range, 2)
approx_three = taylorSinApprox(x_range, 3)
approx_four = taylorSinApprox(x_range, 4)

exact_sin = np.sin(x_range)

err_one = abs(approx_one - exact_sin)
err_two = abs(approx_two - exact_sin)
err_three = abs(approx_three - exact_sin)
err_four= abs(approx_four - exact_sin)
```

```python
plt.plot(x_range, err_one, 'purple')
plt.xlabel("x")
plt.ylabel("Absolute error")
plt.title("n=1")
```

```python
plt.plot(x_range, err_two, 'purple')
plt.xlabel("x")
plt.ylabel("Absolute error")
plt.title("n=2")
```

```python
plt.plot(x_range, err_three, 'purple')
plt.xlabel("x")
plt.ylabel("Absolute error")
plt.title("n=3")
```

```python
plt.plot(x_range, err_four, 'purple')
plt.xlabel("x")
plt.ylabel("Absolute error")
plt.title("n=4")
```

As more terms are added on to the Taylor Series, there is evidently less error for values of x close to 0. However, the error in values farther from 0 become much less accurate, as well. So, essentially, more terms decreases error in low values of x but increases error in high values of x (although higher n makes this explosion of inaccuracy happen later). To prove this, I'll compare n=2 and n=4 with a smaller domain for visual purposes.

```python
plt.plot(x_range, err_two, 'red', label='n=2')
plt.plot(x_range, err_four, 'blue', label='n=4')
plt.xlim(0,1)
plt.ylim(0,0.01)
plt.xlabel("x")
plt.ylabel("Absolute error")
plt.legend()
plt.title("Small values of x")
```
