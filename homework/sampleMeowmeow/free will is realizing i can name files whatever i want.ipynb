```python
import numpy as np
import matplotlib.pyplot as plt
```

A function that calculates Monte Carlo approximation of pi based on an inputted number of samples

```python
def approximatePi(numSamples):
    counter = 0
    for s in range(numSamples):
        x = np.random.uniform(0, 2) - 1
        y = np.random.uniform(0, 2) - 1
        
        if (x**2+y**2 < 1):
            counter = counter + 1
    
    return 4.0 * counter / numSamples
```

Creates an array of data that finds the error in each approximation
A for loop isn't particularly efficient, but who cares! 
I'm using a supercomputer!

```python
errorSet = []
realPi = np.pi
numSampleSet = np.arange(10, 10030, 20)
for i in numSampleSet:
    calculatedPi = approximatePi(i)
    error = abs(calculatedPi - realPi)
    errorSet = np.append(errorSet, error)
```

Plots a graph based on the data above. It's purple.
Purple graph so cute

```python
plt.plot(numSampleSet, errorSet, 'purple')
plt.xlabel("Number of samples in Monte Carlo")
plt.ylabel("Error")
plt.show()
```
